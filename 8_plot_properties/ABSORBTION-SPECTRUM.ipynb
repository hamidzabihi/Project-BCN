```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss_function(a, mu, sigma, min_x, max_x, npoints):
    
    pre_fact = a*1/sigma/np.sqrt(2*np.pi)
    x_grid = np.linspace(min_x, max_x, npoints)
    y = pre_fact*np.exp(-1/2/np.square(sigma)*np.square(x_grid-mu))
    
    return x_grid, y
```

```python
   
##########################################################################################
x = np.loadtxt('out-BCN-V1.log')
len(x[:,1])
a_sum = np.zeros((1,1000))
for i in range(len(x[:,0])):
    
    x_grid, a = gauss_function(x[i,1], x[i,0], 0.05, 0, 6, 1000)
    a_sum += a

    
##########################################################################################
x = np.loadtxt('out-BCN-V1_B_vacancy-TDDFT-PBE.log')
len(x[:,1])
b_sum = np.zeros((1,1000))
for i in range(len(x[:,0])):
    
    x_grid, b = gauss_function(x[i,1], x[i,0], 0.05, 0, 6, 1000)
    b_sum += b

    
##########################################################################################
x = np.loadtxt('out-BCN-V1_C_vacancy-TDDFT-PBE.log')
len(x[:,1])
c_sum = np.zeros((1,1000))
for i in range(len(x[:,0])):
    
    x_grid, c = gauss_function(x[i,1], x[i,0], 0.05, 0, 6, 1000)
    c_sum += c
###########################################################################################
x = np.loadtxt('out-BCN-V1_N_vacancy-TDDFT-PBE.log')
len(x[:,1])
d_sum = np.zeros((1,1000))
for i in range(len(x[:,0])):
    
    x_grid, d = gauss_function(x[i,1], x[i,0], 0.05, 0, 6, 1000)
    d_sum += d
    
###########################################################################################
#x = np.loadtxt('D1T.log')
#len(x[:,1])
#e_sum = np.zeros((1,1000))
#for i in range(len(x[:,0])):
#    
#    x_grid, e = gauss_function(x[i,1], x[i,0], 0.05, 0, 6, 1000)
#    e_sum += e
#
    
    
    

    
figure = plt.figure(num=None, figsize=(3.21, 2.41), dpi=600, edgecolor='black', frameon=True)        
    
#plt.figure()
#plt.plot(x_grid, a_sum.T ,label='Pristine BCN', color='blue', linewidth=2)
#print(x_grid)
#print(a_sum.T)
#print(b_sum.T)
#print(c_sum.T)
#print(d_sum.T)
np.savetxt("b_sum.T.txt", b_sum.T, delimiter='  ')
np.savetxt("c_sum.T.txt", c_sum.T, delimiter='  ')
np.savetxt("d_sum.T.txt", d_sum.T, delimiter='  ')

#plt.plot(x_grid, b_sum.T ,label='BCN with B vacancy',  linewidth=2, color='green')
#plt.plot(x_grid, c_sum.T ,label='BCN with C vacancy', linewidth=2, color='cyan')
#plt.plot(x_grid, d_sum.T ,label='BCN with N vacancy', linewidth=2, color='red')
#plt.plot(x_grid, e_sum.T ,label='D1T',  linewidth=5, color='orange')

plt.legend(fontsize=5.75, ncol=1, loc='upper right')
plt.title('Absorbance')
plt.xlabel('Energy, eV')
# plt.ylabel('Oscillator strength, a.u.')
plt.ylabel('Intensity (a.u.)')
plt.yticks([])
plt.xlim(0, 3)
plt.tight_layout()
plt.savefig('ABSORBTION OF BCN',dpi=600)

plt.show()
```
